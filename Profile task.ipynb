#### Обработка данных:

Особой обработки данных не требуется, кроме стандартизации. Возможно, в дальнейшем может потребоваться анализ признаков на взаимозависимости.

```python
from sklearn import preprocessing
```

```python
scaler_std = preprocessing.StandardScaler()
df.iloc[:, 1:] = scaler_std.fit_transform(df.iloc[:, 1:])
df.head()
```

#### Методы машинного обучения

Для решения задачи бинарной классификации стоит обучить несколько методов машинного обучения:
1. Алгоритм Бернулли наивного байесовского классификатора;
2. Логистическая регрессия;
3. Деревья решений;
4. SVM-классификатор;
5. Случайный лес;
6. Градиентный бустинг;
7. Нейронная сеть.

Для разбиения выборки возьмем процентное соотношение обучающей, валидационной и тестовой выборки соответственно как 65%, 15%, 20%.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
x = df.iloc[:, 1:].values
y = df.iloc[:, :1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.65, stratify=y)
```

Следовательно, на данный момент в тестовой выборке:
- 35% - 100%
- 15% -> 42%

Следовательно, валидационная выборка занимает 42%.

```python
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, train_size = 0.42, stratify=y_test, random_state=42)
```

И поскольку данные для Y являются двумерным массивом, то преобразуем в одномерный.

```python
y_train = y_train.ravel()
y_val = y_val.ravel()
y_test = y_test.ravel()
```

Будем сравнивать модели по основным метрикам для классификации, особое внимание уделяя ROC-AUC, при помощи функции:

```python
def rocauc(model1, y_test, x_test):
    roc_auc = roc_auc_score(y_test, model1.predict_proba(x_test)[:, 1])
    print(f'ROC-AUC Score: {roc_auc:.5f}')
    print('-------------------------------------------------------------')
    print(f'Accuracy: {accuracy_score(y_test, model1.predict(x_test)):.3f}')
    print(f'Precision: {precision_score(y_test, model1.predict(x_test)):.3f}')
    print(f'Recall: {recall_score(y_test, model1.predict(x_test)):.3f}')
    print(f'F1-Score: {f1_score(y_test, model1.predict(x_test)):.3f}')
```

### Алгоритм Бернулли наивного байесовского классификатора

Наивный байесовский алгоритм — это вероятностный алгоритм машинного обучения, основанный на применении теоремы Байеса.

Преимущества:
- Алгоритм очень быстр по сравнению с другими алгоритмами классификации.
- Поскольку набор данных небольшой, то алгоритм обычно дает более точные результаты по сравнению с другими алгоритмами классификации.

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bernoulli_model = BernoulliNB()
bernoulli_model.fit(x_train, y_train)
```

```python
rocauc(bernoulli_model, y_test, x_test)
```

Преимущества:
- простота;
- не требует специальной подготовки данных.

```python
def MakingTree(crit, alpha):
    model1 = tree.DecisionTreeClassifier(criterion = crit, ccp_alpha = alpha)
    model1.fit(x_train, y_train)

    tree.plot_tree(model1, filled = True, rounded = True)
    plt.savefig('tree.pdf') ##для сохранение в пдф
    plt.show

    rocauc(model1, y_test, x_test)
```

### SVM-классификатор

Метод опорных векторов (англ. support vector machine, SVM) — один из наиболее популярных методов обучения, который применяется для решения задач классификации и регрессии. Основная идея метода заключается в построении гиперплоскости, разделяющей объекты выборки оптимальным способом.

Преимущества:
- более уверенная классификация;
- хорошо работает с пространством признаков большого размера;
- хорошо работает с данными небольшого объема.

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix
```

```python
gammas = ['scale', 'auto', 1, 10, 0.1, 20, 0.01]
coef0s = [0, 1, 2, 5, 10, 0.1, 0.01]
degrees = [2, 3, 4, 5]
Cs = [0.001, 0.01, 0.1, 0.5, 1, 2, 10, 20]
```

Для **линейного**:

```python
krl ='linear'

best = [0]
for gamma in gammas:
    for coef0 in coef0s:
        for degree in degrees:
            for C in Cs:
                clf = svm.SVC(kernel = krl, gamma = gamma, coef0 = coef0,
                              degree = degree, C = C, probability=True)
                clf.fit(x_train, y_train)
                ra = roc_auc_score(y_val, clf.predict_proba(x_val)[:, 1])
                if ra > best[0]:
                    best = [ra, gamma, coef0, degree, C]
                        
print(krl, best)

linear_svm = svm.SVC(kernel = krl, gamma = best[1], coef0 = best[2], degree = best[3], C = best[4], probability=True)
linear_svm.fit(x_train, y_train)
rocauc(linear_svm, y_test, x_test)
```

Проверка:

```python
rocauc(linear_svm, y_train, x_train)
```

```python
rocauc(linear_svm, y_val, x_val)
```

Получилась хорошая модель.

Для **сигмоиды**:

```python
krl ='sigmoid'

best = [0]
for gamma in gammas:
    for coef0 in coef0s:
        for degree in degrees:
            for C in Cs:
                clf = svm.SVC(kernel = krl, gamma = gamma, coef0 = coef0,
                              degree = degree, C = C, probability=True)
                clf.fit(x_train, y_train)
                ra = roc_auc_score(y_val, clf.predict_proba(x_val)[:, 1])
                if ra > best[0]:
                    best = [ra, gamma, coef0, degree, C]
                        
print(krl, best)

sigmoid_svm = svm.SVC(kernel = krl, gamma = best[1], coef0 = best[2], degree = best[3], C = best[4], probability=True)
sigmoid_svm.fit(x_train, y_train)
rocauc(sigmoid_svm, y_test, x_test)
```

Проверка:

```python
rocauc(sigmoid_svm, y_train, x_train)
```

```python
rocauc(sigmoid_svm, y_val, x_val)
```

Также неплохая модель.

Для **полинома**:

```python
krl = 'poly'

best = [0]
for gamma in gammas:
    for coef0 in coef0s:
        for degree in degrees:
            for C in Cs:
                clf = svm.SVC(kernel = krl, gamma = gamma, coef0 = coef0,
                              degree = degree, C = C, probability=True)
                clf.fit(x_train, y_train)
                ra = roc_auc_score(y_val, clf.predict_proba(x_val)[:, 1])
                if ra > best[0]:
                    best = [ra, gamma, coef0, degree, C]
                        
print(krl, best)

poly_svm = svm.SVC(kernel = krl, gamma = best[1], coef0 = best[2], degree = best[3], C = best[4], probability=True)
poly_svm.fit(x_train, y_train)
rocauc(poly_svm, y_test, x_test)
```

Проверка:

```python
rocauc(poly_svm, y_val, x_val)
```

Модель имеет хорошие (сбалансированные) метрики.

Для **ядра rbf**:

```python
krl = 'rbf'

best = [0]
for gamma in gammas:
    for coef0 in coef0s:
        for degree in degrees:
            for C in Cs:
                clf = svm.SVC(kernel = krl, gamma = gamma, coef0 = coef0,
                              degree = degree, C = C, probability=True)
                clf.fit(x_train, y_train)
                ra = roc_auc_score(y_val, clf.predict_proba(x_val)[:, 1])
                if ra > best[0]:
                    best = [ra, gamma, coef0, degree, C]
                        
print(krl, best)

rbf_svm = svm.SVC(kernel = krl, gamma = best[1], coef0 = best[2], degree = best[3], C = best[4], probability=True)
rbf_svm.fit(x_train, y_train)
rocauc(rbf_svm, y_test, x_test)
```

### Случайный лес

Случайный лес (random forest) — это множество решающих деревьев. В задаче классификации принимается решение голосованием по большинству. Все деревья строятся независимо.

Преимущества:
- Бэггинг;
- Случайность признака.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
best = [0]
for n in [50, 100, 200, 300, 400, 500, 600]: # количество деревье
    for feat in [3, 5, 7, 10, 12, 15, 17, 20]: # глубина
        for crit in ['gini', 'entropy']: # критерий
            clf = RandomForestClassifier(n_estimators = n, max_features = feat, criterion = crit)
            clf.fit(x_train, y_train)
            ra = roc_auc_score(y_val, clf.predict_proba(x_val)[:, 1])
            if ra > best[0]:
                best = [ra, n, feat, crit]
                        
print("best parameters: ", best)

rfc_model = RandomForestClassifier(n_estimators = best[1], max_features = best[2], criterion = best[3])
rfc_model.fit(x_train, y_train)
rocauc(rfc_model, y_test, x_test)
```

Проверка:

```python
rocauc(rfc_model, y_val, x_val)
```

```python
rocauc(rfc_model, y_train, x_train)
```

Тот факт, что на обучающей выборке все метрики равны 1.0, означает, что случайный лес переобучился под тренировочную выборку.

### Нейронная сеть (многослойный персептрон)

Многослойные персептроны – это типы нейронных сетей, которые являются двунаправленными, поскольку они предваряют распространение входных данных и обратное распространение весов.

Преимущества:
- Возможность изучать нелинейные модели;
- Возможность изучения моделей в режиме реального времени.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

```python
mlp_param_grid = {
    'hidden_layer_sizes': [(50, 25), (50, 50), (100, 50), (100, 100)],
    'activation': ['relu', 'logistic'],
    'alpha': [0.0001, 0.01],
    'learning_rate': ['constant', 'adaptive'],
    'max_iter': [300, 500]
}
```

```python
mlp = MLPClassifier(random_state=42)
mlp_grid_search = GridSearchCV(estimator=mlp, param_grid=mlp_param_grid, cv=3, scoring='roc_auc')
mlp_grid_search.fit(x_train, y_train)
```

```python
mpl_best_params = mlp_grid_search.best_params_
print(f'Лучшие параметры для многослойного персептрона: {mpl_best_params}')
```

```python
mlp_best_model = mlp_grid_search.best_estimator_
rocauc(mlp_best_model, y_test, x_test)
```

Проверка:

```python
rocauc(mlp_best_model, y_val, x_val)
```

```python
rocauc(mlp_best_model, y_train, x_train)
```

Заметно, что на обучающей выборке многослойный персептрон переобучился.
